## **Comparision of Various Differentiation Methods**

```python
import autograd.numpy as np
from autograd import grad
import math
```

```python
def my_function(x):
  return np.sin(x)
```

```python
gradient_values = [] # Initialize an empty list to store the gradient of my_function
x_values = [] # Initialize an empty list to store the x values
cosine_values = [] # Initialize an empty list to store the cosine of the x values
derivative_values = [] # Initialize an empty list to store the numerical approximation of the derivative of sin(x)
step_size = 0.5 # Set the step size for the numerical approximation of the derivative

for x in range(1001):
  x_values.append(((8*math.pi)/1000)*x) # Append the x value to list x_values
  gradient_func = grad(my_function) # Get the gradient function of my_function
  gradient_values.append(gradient_func(((8*math.pi)/1000)*x)) # Append the gradient of my_function at x to list gradient_values
  cosine_values.append(math.cos(x_values[x])) # Append the cosine of x to list cosine_values
  derivative_values.append((math.sin(x_values[x] + step_size) - math.sin(x_values[x]))/step_size ) # Append the numerical approximation of the derivative of sin(x) at x to list derivative_values
```

```python
import matplotlib.pyplot as plt
plt.plot(x_values, gradient_values, color = 'blue', label = 'Differentiation by Autograde')
plt.plot(x_values, derivative_values, color = 'red', label = 'Differentiation by symbolic')
plt.plot(x_values, cosine_values, color = 'black' ,label = 'Differentiation By Numerical')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Comparing Differentiation for different methods')
plt.legend()
plt.show()
```

## **Finding Minima By Using Method Of Gradient Descent**

Gradient descent is an optimization algorithm used to find the minimum of a function. It works by iteratively adjusting the input values in the opposite direction of the gradient of the function at the current point. The size of the steps taken in each iteration is determined by the learning rate, a hyperparameter that controls how quickly the algorithm converges to the minimum.

The basic idea behind gradient descent is that the gradient of a function at a point gives the direction of steepest ascent at that point. By moving in the opposite direction of the gradient, we can iteratively move towards a local minimum of the function.

Gradient descent is commonly used in machine learning to optimize loss functions and improve model performance. There are several variations of gradient descent, including batch gradient descent, stochastic gradient descent, and mini-batch gradient descent, which differ in how they calculate and use the gradient information.

```python
################## For Function x^2 + Y^2 ###################

# Define the function to minimize
def f(x, y):
    return x**2 + y**2

# Define the gradient of the function
def grad_f(x, y):
    return [2*x, 2*y]

# Set the initial point
x, y = 1.0, 1.0

# Set the learning rate
learning_rate = 0.1

# Set the number of iterations
num_iterations = 100

# Perform gradient descent
for i in range(num_iterations):
    # Calculate the gradient at the current point
    grad_x, grad_y = grad_f(x, y)
    
    # Update the point using the gradient descent rule
    x -= learning_rate * grad_x
    y -= learning_rate * grad_y
    
# Print the final point and function value
print(f'f(x,y) = {f(x,y):.4f}')
```
