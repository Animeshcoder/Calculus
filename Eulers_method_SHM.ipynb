```python
# Importing required modules
import math
import matplotlib.pyplot as plt
```

```python
# Define initial values
x0 = 1.0    # initial position
v0 = 0.0    # initial velocity
t0 = 0.0    # initial time

# Define constants
k = 1.0     # spring constant
m = 1.0     # mass
omega = (k/m) ** 0.5
A = 1.0
```

```python
# Define time step and number of steps
dt = 0.01   # time step size
tn = 10
n = int((tn - t0) / dt)    # number of steps

# Set initial values
x = [x0]
v = [v0]
t = [t0]
```

```python
# Loop through each time step and use Euler method to calculate new values
for i in range(1, n):
    x_next = x[-1] + v[-1] * dt    # Euler method for position
    v_next = v[-1] + (-1 * omega ** 2 * x[-1]) * dt   # Euler method for velocity
    t_next = t[-1] + dt
    t.append(t_next)
    x.append(x_next)
    v.append(v_next)
```

```python
# Analytical equation for solution of differential equation of shm.
val = []
def func():
  for i in t:
    val.append(A * math.cos(omega * i))
  return val
```

```python
# Plot position values over time
plt.figure()
plt.plot(t, func(), label = ' Analytical method Graph')
plt.plot(t, x, label = 'Eulers method Graph')
plt.ylabel('position')
plt.xlabel('Time')
plt.legend()
plt.title('Eulers Method Solution')
plt.show()
```

```python
# Access the value of Error
err = []
p = func()
for i in range(len(t)):
  err.append(abs(p[i] - x[i])/abs(p[i]))
```

```python
# Plot the graph of Error
plt.figure(1)
plt.title('Plot of Error')
plt.plot(t, err)
plt.show()
```
