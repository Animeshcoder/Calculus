```python
# Importing required modules
import math
import matplotlib.pyplot as plt
```

```python
# Define the differential equation: y' = y
def f(t, y):
    return y
```

```python
# Define initial conditions
t = [0]
y = [1]

# Define step size and final value of t
dt = 0.1
tn = 5
```

```python
# Applying mid-point method
while t[-1] < tn:
    ti = t[-1] + dt
    yi = y[-1] + dt*f(t[-1] + dt/2, y[-1] + dt/2*f(t[-1], y[-1]))
    t.append(ti)
    y.append(yi)
```

```python
# Analytical Exponential function
def func():
  p = []
  for i in t:
    p.append(math.exp(i))
  return p
```

```python
# Plot the results
plt.figure()
plt.plot(t, y, label = 'Graph of Mid-Point Method')
plt.plot(t, func(), label = ' Graph of Analytical method')
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.title('Midpoint Method Solution')
plt.show()
```

```python
# Access the value of Error
err = []
p = func()
for i in range(len(t)):
  err.append(abs(p[i] - y[i])/abs(p[i]))
```

```python
# Plot the graph of Error
plt.figure(1)
plt.title('Plot of Error')
plt.plot(t, err)
plt.show()
```
