**This code calculates the numerical approximation of the perimeter of an ellipse for different combinations of values for its semi-major and semi-minor axes using numerical integration.**

````python
import math
from scipy.integrate import quad

# Define the function to be integrated
def function(x, i, j):
    return (i/10**2*(math.sin(x))**2 + j/10**2*(math.cos(x))**2)**0.5

# Create lists of values for the semi-major axis a and the semi-minor axis b of the ellipse
a = [i/10 for i in range(20, 50)]
b = [j/10 for j in range(10, 20)]

# Initialize lists to store results
approx_peri_1 = []
approx_peri_2 = []
approx_peri_3 = []
approx_peri_4 = []
error_list = []
f_list = []
a_list = []
b_list = []

# Set value of j
j = 10

# Iterate over all values of i
for i in range(20, 50):
    # Calculate the numerical approximation of the integral using the quad function
    result, error = quad(function, 0, 2*math.pi, args=(i,j))
    
    # Store results
    a_list.append(i/10)
    b_list.append(j/10)
    f_list.append(result)
    
    # Calculate approximations of the perimeter using different formulas
    approx_peri_1.append((abs(result - 2*math.pi*(i/20 + j/20))/result))
    approx_peri_2.append((abs(result - 2*math.pi*((i**2/20 + j**2/20)**0.5))/result))
    approx_peri_3.append((abs(result - math.pi*(6*i/50 + 3*j/40))/result))
    approx_peri_4.append((abs(result - math.pi*(3*(i/10+j/10) - ((3*i/10 + j/10)*(i/10+3*j/10))**0.5))/result))

# Print results
print(f_list)
print(approx_peri_1)
print(approx_peri_2)
print(approx_peri_3)
print(approx_peri_4)
````

````python
import math
from scipy.integrate import quad

# Define the function to be integrated
def function(x, i, j):
    return (i**2*(math.sin(x))**2 + j**2*(math.cos(x))**2)**0.5

# Create lists of values for the semi-major axis a and the semi-minor axis b of the ellipse
a = [i/100 for i in range(20, 50)]
b = [j/100 for j in range(10, 20)]

# Initialize lists to store results
f_list = []
a_list = []
b_list = []

# Iterate over all combinations of a and b
for i in a:
    for j in b:
        # Calculate the numerical approximation of the integral using the quad function
        result, error = quad(function, 0, 2*math.pi, args=(i,j))
        
        # Store results
        a_list.append(i)
        b_list.append(j)
        f_list.append(result)
````

This code calculates the numerical approximation of the perimeter of an ellipse using the quad function from the scipy.integrate module. The perimeter of an ellipse is given by the integral 

```
P = 4 * integral from 0 to pi/2 of sqrt(a^2*sin^2(x) + b^2*cos^2(x)) dx
```
where a and b are the semi-major and semi-minor axes of the ellipse, respectively.

The code defines a function $$function(x, i, j)$$ that represents the integrand of this integral. It takes as arguments the variable of integration x and the values of a and b, represented by i and j, respectively.

The code then creates two lists of values for a and b, represented by the variables a and b. These lists contain values in the range [0.2, 0.5) and [0.1, 0.2), respectively, with a step size of 0.01.

The code initializes three lists to store the results: f_list, a_list, and b_list. It then iterates over all combinations of values in the lists a and b using two nested for loops. For each combination of values, it calculates the numerical approximation of the integral using the quad function from the scipy.integrate module. The result of the integral is stored in the list f_list, while the values of a and b are stored in the lists a_list and b_list, respectively.

````python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd

# Create a new figure
fig = plt.figure()

# Add a 3D axis to the figure
ax = fig.add_subplot(111, projection='3d')

# Create a DataFrame from the data
df = pd.DataFrame({'a': a_list, 'b': b_list, 'f': f_list})

# Plot the data
ax.scatter(df['a'], df['b'], df['f'])

# Set axis labels
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('f')

# Show the plot
plt.show()
````
