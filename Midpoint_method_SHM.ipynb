```python
# Importing required modules
import matplotlib.pyplot as plt
import math
```

```python
# Define initial values
x0 = 1.0    # initial position
v0 = 0.0    # initial velocity
t0 = 0.0    # initial time
```

```python
# Define constants
k = 1.0     # spring constant
m = 1.0     # mass
omega = (k/m) ** 0.5
A = 1.0
```

```python
# Define time step and number of steps
dt = 0.01   # time step size
tn = 10
n = int((tn - t0) / dt)    # number of steps
```

```python
# Define function for SHM motion
def shm(x, omega = (k/m) ** 0.5):
    a = -1 * omega ** 2 * x  # acceleration
    return a
```

```python
# Set initial values
x = [x0]
v = [v0]
t = [t0]
```

```python
# Applying mid-point method
while t[-1] < tn:
    func = shm(x[-1], omega = omega)
    xi = x[-1] + (v[-1] + func * dt / 2) * dt
    vi = v[-1] + func * dt
    ti = t[-1] + dt
    t.append(ti)
    x.append(xi)
    v.append(vi)
```

```python
# Analytical equation for solution of differential equation of shm.
val = []
def func():
  for i in t:
    val.append(A * math.cos(omega * i))
  return val
```

```python
# Plot position values over time
plt.figure()
plt.plot(t, func(), label = ' Analytical method Graph')
plt.plot(t, x, label = 'Mid-Point method Graph')
plt.ylabel('Position')
plt.xlabel('Time')
plt.legend()
plt.title('Mid-Point Method Solution')
plt.show()
```

```python
# Access the value of Error
err = []
p = func()
for i in range(len(t)):
  err.append(abs(p[i] - x[i])/abs(p[i]))
print(err)
```

```python
# Plot the graph of Error
plt.figure(1)
plt.title('Plot of Error')
plt.plot(t, err)
plt.show()
```
