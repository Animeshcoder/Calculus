## **Riemann sum**
A Riemann sum is a method for approximating the area under a curve by dividing the region into smaller rectangles and summing their areas. The intuition behind it is that if we divide the area into very small rectangles, we can calculate the area of each rectangle and then add them to find the area of the total region.

As the number of rectangles gets larger and their width gets smaller, the value of the Riemann sum approximates the area under the curve better. The Riemann sum can then be represented by an integral and is called the Riemann integral of a function from a to b.

Here’s an example:
we want to approximate the definite integral of $$f(x) = x^2 + 2x + 1$$ on the interval $$[0, 1]$$ using a Riemann sum with 4 rectangles. We can divide the interval into 4 subintervals of equal width: $$[0, 0.25], [0.25, 0.5], [0.5, 0.75], and [0.75, 1].$$ Then we can use one of several methods to determine the height of each rectangle.

For example, using the Left Hand Rule, we evaluate the function at the left endpoint of each subinterval and make the rectangle that height. So our Riemann sum would be:

$$f(0) * 0.25 + f(0.25) * 0.25 + f(0.5) * 0.25 + f(0.75) * 0.25 = (0^2 + 20 + 1) * 0.25 + (0.25^2 + 20.25 + 1) * 0.25 + (0.5^2 + 20.5 + 1) * 0.25 + (0.75^2 + 20.75 + 1) * 0.25 = (1) * 0.25 + (1.3125) * 0.25 + (1.75) * 0.25 + (2.3125) * 0.25 = 1.96875$$

So our Riemann sum approximation of the definite integral of $$f(x) = x^2 + 2x + 1 $$on $$[0,1]$$ using 4 rectangles is approximately equal to $$1.96875.$$

```python
############################# Code For Reimann Sum ###################################

def riemann_sum(function, lower_bound, upper_bound, num_of_intervals):
    """
    This function calculates the Riemann sum approximation of the integral of a function.
    :param function: The function to integrate.
    :param lower_bound: The lower bound of the integral.
    :param upper_bound: The upper bound of the integral.
    :param num_of_intervals: The number of intervals to use in the approximation.
    :return: The Riemann sum approximation of the integral.
    """
    interval_width = (upper_bound - lower_bound) / num_of_intervals
    sum = 0
    for i in range(num_of_intervals):
        # Calculate the value of the function at the left endpoint of each interval
        sum += function(lower_bound + i * interval_width)
    return interval_width * sum

print("Riemann sum approximation:", riemann_sum(function, lower_bound, upper_bound, num_of_intervals))
```

## **Trapezoidal rule**
The trapezoidal rule is a technique for approximating the definite integral of a function by approximating the region under the graph of the function as a trapezoid and calculating its area. It is an integration rule used to calculate the area under a curve by dividing the curve into small trapezoids.

Here’s an example:
Let’s say we want to approximate the definite integral of $$f(x) = x^2 + 2x + 1$$ on the interval $$[0, 1]$$ using the trapezoidal rule with 4 trapezoids. We can divide the interval into 4 subintervals of equal width: $$[0, 0.25], [0.25, 0.5], [0.5, 0.75], and [0.75, 1].$$ Then we can use the trapezoidal rule formula to approximate the definite integral:

$$∫f(x)dx ≈ (Δx/2) * [f(x0) + 2f(x1) + 2f(x2) + … + 2f(xn-1) + f(xn)]$$

where $$Δx$$ is the width of each subinterval and $$f(xi)$$ is the value of the function at each endpoint.

In our example, $$Δx = (1 - 0)/4 = 0.25$$ and our trapezoidal rule approximation would be:

$$(0.25/2) * [f(0) + 2f(0.25) + 2f(0.5) + 2f(0.75) + f(1)] = (0.125) * [(0)^2 + 20 + 1 + 2(0.25)^2 + 20.25 + 1 + 2(0.5)^2 + 20.5 + 1 + 2(0.75)^2 + 20.75 + 1 + (1)^2 + 21 + 1] = (0.125) * [1 + 1.3125 + 1.75 + 2.3125 + 4] = (0.125) * [10.375] = 2.34375$$

So our trapezoidal rule approximation of the definite integral of $$f(x) = x^2 + 2x + 1 on [0,1]$$ using 4 trapezoids is approximately equal to $$2.34375.$$

```python
############################# Code For Trapezoidal Rule ###################################

def trapezoidal_rule(function, lower_bound, upper_bound, num_of_intervals):
    """
    This function calculates the trapezoidal rule approximation of the integral of a function.
    :param function: The function to integrate.
    :param lower_bound: The lower bound of the integral.
    :param upper_bound: The upper bound of the integral.
    :param num_of_intervals: The number of intervals to use in the approximation.
    :return: The trapezoidal rule approximation of the integral.
    """
    interval_width = (upper_bound - lower_bound) / num_of_intervals
    # Add half of the value at the endpoints to the sum
    sum = 1/2 * (function(lower_bound) + function(upper_bound))
    for i in range(1, num_of_intervals):
        # Add the value at each interior point to the sum
        sum += function(lower_bound + i * interval_width)
    return interval_width * sum

print("Trapezoidal rule approximation:", trapezoidal_rule(function, lower_bound, upper_bound, num_of_intervals))
```

## **Simpson’s rule**
Simpson’s rule is a method for approximating definite integrals of functions. It is usually more accurate than approximations using Riemann sums or the trapezoidal rule and is exact for linear and quadratic functions1.

Here’s an example: Let’s say we want to approximate the definite integral of $$f(x) = x^2 + 2x + 1$$ on the interval $$[0, 1]$$ using Simpson’s rule with 4 subintervals. We can divide the interval into 4 subintervals of equal width: $$[0, 0.25], [0.25, 0.5], [0.5, 0.75], and [0.75, 1].$$ Then we can use Simpson’s rule formula to approximate the definite integral:

$$∫f(x)dx ≈ (Δx/3) * [f(x0) + 4f(x1) + 2f(x2) + … + 4f(xn-1) + f(xn)]$$

where $$Δx$$ is the width of each subinterval and $$f(xi)$$ is the value of the function at each endpoint.

In our example, $$Δx = (1 - 0)/4 = 0.25$$ and our Simpson’s rule approximation would be:

$$(0.25/3) * [f(0) + 4f(0.25) + 2f(0.5) + 4f(0.75) + f(1)] = (0.0833333) * [(0)^2 + 20 + 1 + 4(0.25)^2 + 20.25 + 1 + 2(0.5)^2 + 20.5 + 1 + 4(0.75)^2 + 20.75 + 1 + (1)^2 + 21 + 1] = (0.0833333) * [1 + 41.3125 + 31.75 + 4*2.3125 + 4] = (0.0833333) * [13] = 2.333$$

So our Simpson’s rule approximation of the definite integral of $$f(x) = x^2 + 2x + 1, in, [0,1]$$ using 4 subintervals is approximately equal to $$2.333.$$

```python
############################# Code For Simpsons Rule ###################################

def simpsons_rule(function, lower_bound, upper_bound, num_of_intervals):
    """
    This function calculates the Simpson's rule approximation of the integral of a function.
    :param function: The function to integrate.
    :param lower_bound: The lower bound of the integral.
    :param upper_bound: The upper bound of the integral.
    :param num_of_intervals: The number of intervals to use in the approximation.
    :return: The Simpson's rule approximation of the integral.
    """
    interval_width = (upper_bound - lower_bound) / num_of_intervals
    # Add the value at the endpoints to the sum
    sum = function(lower_bound) + function(upper_bound)
    for i in range(1, num_of_intervals):
        if i % 2 == 1:
            # Add four times the value at each odd interior point to the sum
            sum += 4 * function(lower_bound + i * interval_width)
        else:
            # Add two times the value at each even interior point to the sum
            sum += 2 * function(lower_bound + i * interval_width)
    return (interval_width / 3) * sum

print("Simpson's rule approximation:", simpsons_rule(function, lower_bound, upper_bound, num_of_intervals))
```

```python
# Values for plot
n = [5*i for i in range(1,100)]
reimann_list = []
trapezoidal_list = []
simpsons_list = []
for i in n:
  reimann_list.append(result - riemann_sum(function, lower_bound, upper_bound, i))
  trapezoidal_list.append(result - trapezoidal_rule(function, lower_bound, upper_bound, i))
  simpsons_list.append(result - simpsons_rule(function, lower_bound, upper_bound, i))
```

```python
import matplotlib.pyplot as plt
plt.plot([1/(upper_bound - lower_bound)*d for d in n], reimann_list)
plt.xlabel("1/dx")
plt.ylabel("Result")
plt.title('Reimans Sum for Integration')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1/(upper_bound - lower_bound)*d for d in n], trapezoidal_list)
plt.xlabel("1/dx")
plt.ylabel("Result")
plt.title('Trapezoidal rule for Integration')
plt.show()
```
