```python
# Importing required modules
import math
import matplotlib.pyplot as plt
```

```python
# Define the step size t
dt = 0.1

# Define the initial conditions:
t0 = 0
y0 = 1
```

```python
# Set the range of values for the independent variable
tn = 5
n = int((tn - t0) / dt)
```

```python
# Applying Eulers method
t = [t0]
y = [y0]

for i in range(n):
    y_next = y[-1] + dt * y[-1]
    t_next = t[-1] + dt
    y.append(y_next)
    t.append(t_next)
```

```python
# Analytical Exponential function
def fun1():
  p = []
  for i in t:
    p.append(math.exp(i))
  return p
```

```python
# Plot of both graphs
plt.figure()
plt.plot(t, y, label = 'Graph of Eulers Method')
plt.plot(t, fun1(), label = ' Graph of Analytical method')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.title('Eulers Method Solution')
plt.show()
```

```python
# Access the value of Error
err = []
p = fun1()
for i in range(len(t)):
  err.append(abs(p[i] - y[i])/abs(p[i]))
```

```python
# Plot the graph of Error
plt.figure(1)
plt.title('Plot of Error')
plt.plot(t, err)
plt.show()
```
